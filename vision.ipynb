```python
import cv2 as cv
import matplotlib.pyplot as plt
from ultralytics import YOLO
import glob
import time
```

```python
cam = cv.VideoCapture(0)
model = YOLO('yolov8n.pt')
```

```python
lista_memes = []
# buscamos todos los archivos en la carpeta imagenes
archivos = glob.glob('imagenes/*')
for ruta in archivos:
    # AQUÍ usamos cv.imread para convertir el archivo en una matriz de píxeles
    img = cv.imread(ruta)
    if img is not None:
        # redimensionamos para que todos tengan el mismo tamaño
        img = cv.resize(img, (400, 400))
        lista_memes.append(img)
```

```python
import cv2 as cv
import time
import random


indice_meme = 0
ultimo_cambio = 0
meme_actual = None
ventana_meme_abierta = False

while True:
    ret, frame = cam.read()
    if not ret: break

    result = model(frame, stream=True)
    persona, movil = [], []

    for r in result:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cls = int(box.cls[0])
            if cls == 0: persona.append((x1, y1, x2, y2))
            elif cls == 67: movil.append((x1, y1, x2, y2))

    detectado = False 
    for px1, py1, px2, py2 in persona:
        for mx1, my1, mx2, my2 in movil:
            cmx, cmy = (mx1 + mx2)//2, (my1 + my2)//2
            if (px1 < cmx < px2) and (py1 < cmy < py2):
                detectado = True

    if detectado and len(lista_memes) > 0:
        tiempo_ahora = time.time()
        
        if (tiempo_ahora - ultimo_cambio > 2.0) or (meme_actual is None):
            
            # si hay una ventana abierta, la cerramos para que se "refresque"
            if ventana_meme_abierta:
                cv.destroyWindow("ALERTA_MEME")
            
            # elegimos el siguiente meme
            meme_actual = lista_memes[indice_meme]
            indice_meme = (indice_meme + 1) % len(lista_memes)
            ultimo_cambio = tiempo_ahora
            
            # creamos la ventana nueva
            cv.imshow("ALERTA_MEME", meme_actual)
            

            cv.moveWindow("ALERTA_MEME", 1000, 100) 
            
            ventana_meme_abierta = True
    else:
        # Si dejas de mirar, cerramos la pestaña a la derecha
        if ventana_meme_abierta:
            cv.destroyWindow("ALERTA_MEME")
            ventana_meme_abierta = False
        meme_actual = None
        ultimo_cambio = 0

    cv.imshow('Camara Principal', frame)
    #movemos la cámara principal a la izquierda para que no se solapen
    cv.moveWindow('Camara Principal', 0, 100)

    if cv.waitKey(1) == ord('q'): break

cam.release()
cv.destroyAllWindows()
```
